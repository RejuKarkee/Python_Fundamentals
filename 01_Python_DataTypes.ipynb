```python
#Slicing
r = "Hello, Rezz"
print(r[2:5])
print(r[5])
print(r[-5:-1])
print(r[-5:])
```

```python
#upper & lower() method
r = "Hello, Rezz!"
print(r.upper())
print(r.lower())
```

```python
#replace
print(r.replace("H","B"))
```

```python
#split()
print(r.split(","))
```

```python
#can combine string and number with format() method
age = 58
txt = "My name is Sally, and I am"
print (f"{txt}  {age}")
```

```python
#format() method takes unlimited number of arguments
quality = 3
itemno = 567
price = 49.95
myorder = "I want {} pieces of item {} for {} dollars"
print(myorder.format(quality, itemno, price))
print(f"I want {quality} pieces of item {itemno} for {price} dollars") #another approach

myorder = "I want {0} pieces of item {1} for {2} dollars" #extracting through indexing
print(myorder.format(quality, itemno, price))
```

```python
#Almost any value is evaluated to True if it has some sort of content.
#Any string is True, except empty strings.
#Any number is True, except 0.
#Any list, tuple, set, and dictionary are True, except empty ones.

bool("abc")
bool(123)
bool(["apple", "cherry", "banana"])
```

```python
# all these value will come out to be False
bool(False)
bool(None)
bool(0)
bool("")
bool(())
bool([])
```
