```python
#case sensitive
a = 4
print(a)

A = "Sally"
print(a)
print(A)
```

```python
#values to multiple variables & also can assign same value to multiple variables
x,y,z = "Orange", "Watermelon", "Strawberries"
print(x)
print(y)
print(z)

x = y = z = "Apple"
print(x)
print(y)
print(z)
```

```python
#unpack a collection: If you have a collection of values in list, tuple etc.
 #Python allows you to extract the values into variables. This is called unpacking
fruits = ["Orange", "Watermelon", "Strawberries"]
x, y, z = fruits
print(x)
print(y)
print(z)
```
