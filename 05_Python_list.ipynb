```python
#list can store multiple items in a single variable
list = ["apple", "bird" , 5 ,8,  6.8, 5]
print(list)
```

```python
print(len(list))
```

```python
#accesing through index number
print(list[2])
```

```python
print(list[2:5])
print(list[:3])
```

```python
#list is mutable
list[1] = "pegion"
print(list)
```

```python
#can use insert() method to insert an item at the specific index
list.insert(1, "Orange")
print(list)
```

```python
#can remove an item as well
list.remove("Orange")
print(list)
```

```python
#can use pop to remove a last item or from specific index.
list.pop()
print(list)
list.pop(1)
print(list)
```

```python
#del removes the entire list or as per assigned index
del list

# clear() method empties the list.
list.clear()
print(list)
```

```python
#can loop in the list as well
```

```python
for x in list:      #this will iterate the value of list
    print(x)
```

```python
for i in range(len(list)):      #will return all items by referring to their index number
    print(list[i])
```

```python
i = 0
while i < len(list):            #will return all items by referring to their index number
    print(list[i])
    i += 1
```

```python
#list comprehension
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []
for x in fruits:
    if "a" in x:
        newlist.append(x)
print(newlist)
```

```python
#sort() the list alphabetically
list1 = ["Rezz", "Shi", "Pii", "Roro"]
list1.sort()
print(list1)
```

```python
#sorting desc
list1 = ["Rezz", "Shi", "Pii", "Roro"]
list1.sort(reverse = True)
print(list1)
```
