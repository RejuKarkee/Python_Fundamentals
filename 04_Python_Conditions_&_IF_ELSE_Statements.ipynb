```python
#elif
for num in range(1,51):
    if num % 3 == 0 and num % 5 == 0:  #and operator works when both the conditions is meet
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```
