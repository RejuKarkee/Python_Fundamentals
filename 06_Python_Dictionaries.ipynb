```python
#Dictionary items are ordered, changeable, and does not allow duplicates.
#Dictionary items are presented in key:value pairs, and can be referred to by using the key name.

dicti = {
    "brand": "Ford",
    "model": "Mustang",
    "year" : 1964
}
print(dicti)
print(dicti["brand"])
```

```python
print(len(dicti)) #return the number of key,value pair or items in the dictionary
```

```python
#can create a dictionary with dict() method
thisdict = dict(name = "John", age = 36, country = "Norway")
print(thisdict)
```

```python
#get keys
car = {
"brand": "Ford",
"model": "Mustang",
"year": 1964
}
x = car.keys()
print(x)
```

```python
#adding a new key value pai inside the dictionary
car["color" ] = "white"
print(x)

x = car.values()
print(x)

#updating the value
car["year"] = 2000
print(x)
```

```python
#get key:value pair
x = car.items()
print(x)
```

```python
#check if model exist in the dictionary
if "model" in car:
    print("Yes, it exist. It is one of the key in the dictionary")
else:
    print("Nahh!!")
```

```python
#pop() method removes the item with the specified key name:
car.pop("model")
print(car)
```

```python
car2 = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
for x in car2:  #print key names
    print(x)

for x in car2:      #print value names
    print(car2[x])
```

```python
for x in car2.values():
    print(x)
```

```python
for x in car2.keys():
    print(x)
```

```python
for x in car2.items():
    print(x)
```

```python
#copy() method
thisdict = {
 "brand": "Ford",
 "model": "Mustang",
 "year": 1964
}
mydict = thisdict.copy()
print(mydict)
```
